We have many different figures that can support our feature selection, and we actually believe most of the visualizations are quite helpful and informative.

![Sex](figures/Sex_plot.png)
<br>
<br>
The first figure we selected was the sex of the passenger, it is quite clear that the female survival rate is much higher than male survival rates, this potentially indicates that females are given priority when getting onto lifeboats. This could potentially be supplemented with a further breakdown, for instance a further breakdown in terms of age brackets. Likely what gave females an advantage should also give children an advantage is they are prioritized on getting onto lifeboats. 


![Pclass](figures/Pclass_plot.png)
<br>
<br>
The second figure we chose is passenger class. Again, it is quite clear that passengers in more expensive classes have higher survival chances. This likely indicates that they are prioritized when getting onto lifeboats, they may also be aware slightly earlier of the boat sinking. If the movie is to be believed, then those with better passenger classes likely also had better rooms that are closer to the service. Unsurprisingly, a variable that echos this is fare interval. those with the highest fare also had the highest survival rate.

![Age](figures/AgeInterval_plot.png)
<br>
<br>
Age is another very important factor. Intuitively, it should be the case that young people are more likely to survive, as they are young and stronger. However, the data does not reflect that, so we must model it carefully, but this actually makes sense and is aligned with out other hypotheses. Interval 0 unsurprisingly has quite a high survival rate, which largely reflects the hypothesis that children are given priority. Those that are older also had a fairly how survival rate, which shows they may have been given priority as well.

Finally, figures surrounding whether the passenger has related passengers contains some useful information. 
<br>
<br>
![sibsp](figures/SibSp_plot.png)
<br>
<br>
Firstly, whether they had sibling or spouses show some interesting data. Those without actually had a significantly lower survival rate, this may be due to several reasons. Firstly, it may be the fact that those who had siblings or spouse on board are relatively more wealthy, so the data is actually reflecting the passenger class. Secondly, those with siblings may indicate that themselves are also younger, so it is actually reflecting the children and woman first policy. 
<br>
<br>
![Parch](figures/Parch_plot.png)
<br>
<br>
Similarly, whether they had parents or children abroad can lead to similar discussions, it is likely that those with children are given priority as it may be the case that they believe someone needs to take care of the children. 
<br>
<br>
![Famsize](figures/FamilySize_plot.png)
<br>
<br>
Grouping all of this together, this links onto the family size figure. It seems clear that those alone (likely workers on board or individuals who cannot afford expensive tickets) have a low survival rate, while to a certain extent, those in smaller families have a higher survival rate possible due to the woman/children first policy. 

It is clear that a lot of figures must be used contextually in our case, so further analysis of interactions between variables are necessary to get a clearer idea.
